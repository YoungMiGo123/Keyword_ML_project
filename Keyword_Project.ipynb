```python
# Importing Lemmatizer library from nltk
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer() 

def ContainsinList(item_list,item):
    first = 0
    last = len(item_list)-1
    found = False
    while( first<=last and not found):
        mid = (first + last)//2
        if item_list[mid] == item :
            found = True
        else:
            if item < item_list[mid]:
                last = mid - 1
            else:
                first = mid + 1
    return found


def cleankeys(listdata):
    result = []
    for i in listdata:
            #print(i)
            if type(i) == str:
                tmp = []
                tokens = word_tokenize(i)
               
                for word in tokens:
                    lemon = lemmatizer.lemmatize(word.lower())
                    if lemon.endswith('-'):
                        lemon = lemon.replace('-','')
                    #print(f'Original = {word.lower()}, Result = {lemon}\n')
                    tmp.append(lemon)
                stopwords = [x for x in tmp if x not in a]
                if 'and/or' in stopwords:
                        stopwords.remove('and/or')
                if '-' in stopwords:
                    stopwords.remove('-')
                if ContainsinList(stopwords,'('):
                    stopwords.remove('(')
                if ContainsinList(stopwords,')'):
                    stopwords.remove(')')
                stopwords = set(stopwords)
                result += stopwords
    return result
            

keywords = data[['Group','Keywords']]
#print(keywords)
bigData = []
dict = {}
tmpId = ''

count = 0    

for row in keywords.itertuples(index=True, name='Pandas'):
    groupId = row[1]
   
    list = row[2].split(', ')
    clean_list = set(cleankeys(list))
    #tags = nltk.pos_tag(clean_list)
    #chunk = ne_chunk(tags)
    #print(chunk)
    print(f'Before -> Group = {groupId}, Keys = {list}\n')
    print(f'After -> Group = {groupId}, Keys = {clean_list}\n\n')
    ##Run line only once
    if count == 0:
        count += 1
        tmpId = row[1]
    ##Once we've cleared a group of items, start with the next group
    if not (groupId == tmpId):
        cleandata = cleankeys(bigData)
        dict[tmpId] = set(cleandata)
        #print(f'Group = {tmpId}, Keywords = {dict[tmpId]}\n')
        #print(dict[tmpId])
        bigData = []
        tmpId = groupId
    
    bigData += clean_list
    
#print(dict)
```
